# Base-map of POI clusters for airspace partitioning

> aka. Airspace infrastructure discovery via POI aggregation

## Core idea

From `processor.py` and the motion_prediction section, I have derived csv files for the POIs:

$$
\mathrm{P} = {(x_i, y_i, z_i, s_i)}
$$

where: $s_i$ = POI-score

My point is:

* **cluster POIs** in space (optionally spaceâ€“time)
* **partition airspace** based on *POI density and intensity*
* explicitly reject isolated POIs as anomalies

This is:

* unsupervised
* statistics-driven
* structural

## Clustering

Based on surfing the Internet:
- Use HDBSCAN or Weighted DBSCAN to deal with *density-aware clustering*
- Or, maybe Kernel density + watershed can be applied to this as well

> **The clustering partition should be 3D!**
