# Base-map of POI clusters for airspace partitioning

> aka. Airspace infrastructure discovery via POI aggregation

## Core idea

From `processor.py` and the motion_prediction section, I have derived csv files for the POIs:

$$
\mathrm{P} = {(x_i, y_i, z_i, s_i)}
$$

where: $s_i$ = POI-score

My point is:

* **cluster POIs** in space (optionally space–time)
* **partition airspace** based on *POI density and intensity*
* explicitly reject isolated POIs as anomalies

This is:

* unsupervised
* statistics-driven
* structural

## ***Perhaps, I should create something new***

We define an attention distribution function (ADF), a score-weighted spatial intensity field derived from detected points of interest (POIs). The ADF represents the semantic importance of airspace regions induced by collective trajectory behavior.

**However, it is not a brand new theory**

It is just a generalized, marked, anisotropic KDE whose application and interpretation are novel.

### Comparison with Standard KDE

A **standard KDE** assumes:

* i.i.d. samples
* no semantics
* no dynamics
* no physics
* usually 2D
* fixed bandwidth
* no interpretation beyond density


| Component      | Classical KDE | Your Model                       |
| -------------- | ------------- | -------------------------------- |
| Dimensionality | 2D            | **3D**                           |
| Weights        | uniform       | **semantic scores**              |
| Kernel         | isotropic     | **adaptive / anisotropic**       |
| Samples        | i.i.d.        | **trajectory-derived**           |
| Interpretation | density       | **attention / importance**       |
| Purpose        | estimation    | **airspace structure discovery** |

### Verdict

> *“The proposed attention distribution function can be interpreted as a semantic-weighted kernel density field, equivalent to the intensity function of a marked spatial point process. While the underlying estimation technique is related to KDE, the novelty lies in its construction from physics-based trajectory deviations and its use for airspace semantic partitioning.”*
