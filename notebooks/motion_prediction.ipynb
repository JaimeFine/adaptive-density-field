# Motion Prediction Method's Contemplation

## How should I predict aircrafts' motion?

- Pure physic-based method
- Data-driven (Not realistic if I want to find poi in the data --- I cannot trust the data)
- ML + physic-based method

### Pure physics-based method:

Whether the prediciton use:
- Only the first previous position
- Several previous position.
- Or add the later position (combining first previous nad first later position)

**If multiple previous position, then the model might turn into a physics-based 3d interpolation**
**Or, it can be used to calculate acceleration**

However, the deviation is still inevitable! But if we use clustering model, we can minimize the effect of deviation and find the ***relative deviation***

> Well, how to find the so-called ***relative deviation***?
>
> I think one way might work well: KNN or DBSCAN algorithm, we cluster the first K nearest neighbours of that trajectory point!

### ML + physic-based method:

Maybe we have to combine the pure physics-based method with ML algorithms:
- LSTM
- Transformer
