## 2. Potential Algorithm

The point is exactly what I have talked about. I have no idea where the POIs are, so I have to use data to predict which position is most likely to be a POI, or is there a POI between two timestamp. So I need a threshold (I might have some data about several POI data). I need timestamp because I need to use time and velocity to calculate the expected location in the next timestamp, and see how far it is between the predicted location and the actual location in the next stamp. And use that to predict POIs! Or it can be understood in this way: it is an interpolation, I mask one dot between two dots, and use regression (not necessarily linear) to predict a route, and how far the masked point is from the route determine the possibility of the existence within that route (not necessarily that masked dot).

### Bisection Prediction

> I didn't apply this method yet, but I think it is worth exploring!

Link the start and end points and apply binary interpolation recursively.
At each step, identify the vertex exhibiting the maximum deviation and increment its score.
By iteratively subdividing and scoring, a scoring matrix is constructed, from which the optimal threshold is derived.

**This method, however, assumes that we have line-route instead of discrete trajectory points.**
